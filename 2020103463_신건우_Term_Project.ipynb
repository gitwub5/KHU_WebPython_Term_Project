#### 주제: 가상화폐 차트 분석

1. 주제 선정 이유: 가상화폐는 시간이 지날수록 더욱 높은 관심을 받는 분야이다. 최근에 수많은 관련 기사가 쏟아져 나올 만큼 큰 이슈인데 이에 남녀노소 할 것 없이 가상화폐 시장에 뛰어드는 추세다. 급여로는 돈을 모으기 힘든 현실인 만큼 가상화폐를 통해 재테크를 하는 것이다. 현재 가상화폐는 도박과 별다르지 않다는 인식이 강하다. 그것은 가상화폐 시장의 알수없는 변동성 때문이다. 주식과는 다르게 회사의 가치를 예상할 수 없을뿐더러 일명 큰손이라고 불리는 큰 가상화폐 투자자들의 영향력에 의해 가상화폐 시장은 알 수 없게 변동하기 때문이다. 따라서, 주식 시장보다는 심리적인 요인에 영향을 더 크게 받는다. 하지만, 이러한 예측이 불가능한 가상화폐 시장에서도 보이는 패턴이 있다. 언제나 들어맞는 패턴이 아닐지라도 분석을 통해 이러한 패턴을 찾고 미래를 예측하여 수익을 창출할 수 있다. 큰 돈을 번 트레이더들 사이에서는 가상화폐의 차트에서 공통적으로 확인하는 지표가 있다. 반대로 트레이더마다 사용하는 전략이 각각 다르다는 점이 있다. 즉, 자신만의 노하우를 통해 수익을 창출하는 것이다. 이번 프로젝트를 통해 코인의 차트 크롤링하여 분석하고 의미가 있는 지표로 변환하는 작업을 해볼 것이다.

2. 주제 정의: 많은 코인 트레이더들이 적용하는 기본적인 개념인 지지와 저항을 차트에 적용해봄으로써 유의미한 지표인지 아닌지 확인해볼 것이다. 지지와 저항이란 코인을 어느 시점에 매수하고 매도해야 하는지 도움을 주는 개념이다. 지지는 투자자들이 심리적으로 매수를 할 만한 구간을 의미하고 저항을 반대로 투자자들이 심리적으로 매도를 할만한 구간을 말한다. 즉, 투자자들의 심리를 차트에 적용하여 안정적인 수익을 기대하는 것이다. 그렇다면 차트에서 지지선과 저항선은 어디에 그어야 하는가? 차트의 상황에 따라 수평인지 아닌지 다르게 적용되지만, 일반적으로는 고점끼리 이은 일직선을 지지선, 저점끼리 이은 일직선을 지지선으로 사용한다. 상승추세인 경우 저점끼리 이은 추세선을 지지선으로 보고, 하락 추세인 경우 고점끼리 이은 추세선을 저항선으로 본다. 아래의 사진 자료와 같은 개념이다.  

    따라서 상승추세 저점에서 매수하고, 하락추세 고점에서 매도를 하는게 지지와 저항의 개념을 이용한 매매법이다. 차트는 언제나 위의 사진처럼 그려지지 않고, 지지선이 저항선으로 변하는 경우도 있고 저항선이 지지선으로 변하는 경우도 발생한다. 


    위의 사진과 같이 상승추세에서 지지선이 무너질 경우 지지선이 일시적으로 저항선의 역할을 하기도 한다. 이렇게 지지선이 무너진 경우 지지선은 저항선으로 변경되며 다시 추세선을 그려 저항선을 분석해야한다. 반대로 하향추세에서 저항선을 돌파해 주가가 상승한 경우 저항선은 지지선으로 변경되며 추세선을 다시 그려 차트를 분석해야한다. 이 개념이 언제나 들어맞는건 아니기 때문에 트레이더들은 더 많은 차트 분석 기술을 이용한다. 가장 많이 사용되는 예시로는 피보나치되돌림이라는 기술이 있다. 이는 상승하는 차트에서 조정되는 비율을 피보나치 수열에 접목시킨 개념으로 다음 조정은 어느 지점에서 끝날지 예측하는 분석 기술이다. 이 개념에도 지지선과 저항선의 개념이 적용되고 지지와 저항의 개념을 한층 더 의미 있게 만들어 주는 기술이다. 
    
    위와 같은 개념을 통해 프로젝트에서 불어온 데이터를 통해 차트를 가공한 후 프로그램이 지지선과 저항선을 자동으로 그릴 수 있게 해볼 것이다. 그리고 지속적으로 차트를 분석하여 이러한 지지와 저항의 개념을 통한 기술들이 의미가 있는지 분석해볼 예정이다.

```python
#선택한 종목의 10분봉 차트 가져오기(캔들 101개)
from pandas import DataFrame
import pandas as pd
import numpy as np

unit = '10' 
count = '101' #최대 200개
market_URL = "https://api.upbit.com/v1/candles/minutes/{0}?market={1}&count={2}".format(unit,market,count)

response2 = requests.get(market_URL)
market_data= response2.json()

#필요한 데이터만 추출(candle_date_time_kst,opening_price,high_price,low_price,trade_price,candle_acc_trade_volume)
def extract_Data(list):
    data = []
    for elem in reversed(list):
        data_elem=[]
        data_elem.append(elem.get("candle_date_time_kst"))
        data_elem.append(elem.get("opening_price"))
        data_elem.append(elem.get("high_price"))
        data_elem.append(elem.get("low_price"))
        data_elem.append(elem.get("trade_price"))
        data_elem.append(elem.get("candle_acc_trade_volume"))
        data.append(data_elem)
    return data

key_data = extract_Data(market_data)

#dataframe 출력
df = DataFrame(key_data,columns=['date','open','high','low','close','volume'])
print(df)
```

3) 저항선 지지선 찾기: 
 먼저 차트의 상승세와 하락세를 구분지을 필요가 있다. 그 다음에 상승세일 경우 지지선을 찾고 하락세일 경우 저항선을 찾아서 차트의 동향을 분석해볼것이다. 상승세일 때 양봉의 각 종가 데이터를 Tensorflow를 사용하여 선형회귀하여 저항선에 가장 가까운 선을 얻을 것이다. 반대로 하락세일 때 음봉의 각 종가 데이터를 사용할 것이다.
 
 방법:
 초반 그래프의 기울기를 통해 파악 -> 상승(지지선),하락(저항선) -> 지지선 돌파 = 하락 추세, 저항성 돌파 = 상승 추세 -> 추세 변화
 (+이동 평균을 사용해 상승장 하락장 판단: 이동평균보다 종가가 높다면 상승장, 그렇지 않다면 하락장)
 
 1.이동 평균선(MA5)의 기울기(평균변화율)를 구한 후 TrendIs1 변수에 양수이면 상승장(True), 음수이면(False) 반환 + 이동평균(MA5)보다 종가가 높다면 TrendIs2 변수에 상승장, 그렇지 않다면 하락장 반환 / 2. 이동평균선(MA2)의 평균변화율 값 구하기
 / 3.상승장인 경우 이동평균선(MA2)에서 기울기가 음(또는 0)에서 양으로 변할 때의 종가와 저가, 하락장인 경우 기울기가 양에서 음(또는 0)으로 변할 때의 시가와 고가를 가져온다. / 4.선형회귀를 tensorflow를 이용하여 구현하고 저항선 또는 지지선의 일차함수를 구한다. / 5.저항선 또는 지지선을 돌파하는 지점부터 반대 추세의 값과 선을 구한다.

```python
#상승세 하락세 구분 함수 (상승세: True 반환, 하락세: False 반환)
def Set_trend(start, df_ma10, df_close): #(start는 9이상)
    TrendIs1 = False
    TrendIs2 = False
    
    if(df_ma10.iloc[start+1] - df_ma10.iloc[start] > 0):
        TrendIs1 = True
    if(df_close.iloc[start] - df_ma10.iloc[start] > 0):
        TrendIs2 = True
        
    if(TrendIs1 and TrendIs2 == True):
        return True
    elif((not TrendIs1) and (not TrendIs2) == True):
        return False
    else: 
        print("※오류 - 데이터 부족(사용자 초기 추세 지정)")
        return custom_trend()
    
#market_trend 함수로 시장 추세를 판단하기 어려운 경우 사용자가 직접 추세 지정
def custom_trend(): 
    while True:
        TrendIs = int(input("1.상승장\n2.하락장\n입력하시오: "))
        if(TrendIs == 1):
            return True
        elif(TrendIs == 2):
            return False
        else:
            print("잘못 입력하셨습니다. 다시 입력해주세요.")

#초기 그래프 추세
Trend = Set_trend(9, df['close'].rolling(window=5).mean(), df['close'])

print("초기 시장 추세:", end=' ')
if Trend: 
    print("상승장")
else: 
    print("하락장")
    
#추세 바뀌는 지점 지정(처음에만)
def end_trend(start, trend, df_ma20, df_close): #(start는 19이상)
    end_index = len(df_ma20)
    if(trend):
        for i in range(start,len(df_ma20)):
            if(df_ma20.iloc[i] > df_close.iloc[i]):
                end_index = i
                break
    else:
        for i in range(start,len(df_ma20)):
            if(df_ma20.iloc[i] < df_close.iloc[i]):
                end_index = i
                break
    return end_index

start_index = 1
end_index = end_trend(19, Trend, df['close'].rolling(window=20).mean(), df['close'])
```

```python
#런타임:2분~3분
#기울기 반환 함수
def get_gradient(ma):
    list = []
    for i in range(len(ma)-1):
        if(ma.iloc[i] == np.NaN):
            list.append(np.NaN)
        list.append(ma.iloc[i+1] - ma.iloc[i])
    line_df = DataFrame(list, columns=['gradient'])
    line_df['gradient'] = pd.to_numeric(line_df['gradient'])
    return line_df

Df_d = get_gradient(df['close'].rolling(window=2).mean())

#기울기가 음(또는 0)에서 양으로 변할 때
def data_uptrend(start, end, df_d, df_open, df_low):
    list = []
    list_data = []
    for i in range(start, end): #(start는 1이상)
        if(df_d.iloc[i].iloc[0]<= 0 and df_d.iloc[i+1].iloc[0] > 0):
            list.append(i+1)
    for i in range(len(df_open)):
        get_data = []
        if(i in list):
            get_data.append(df_open.iloc[i])
            get_data.append(df_low.iloc[i])
            list_data.append(get_data)
            continue
        get_data.append(np.NaN)
        get_data.append(np.NaN)  
        list_data.append(get_data)
    data = DataFrame(list_data, columns=['open','low'])
    return data

#기울기가 양에서 음(또는 0)으로 변할 때
def data_downtrend(start, end, df_d, df_open, df_high):
    list = []
    list_data = []
    for i in range(start, end): #(start는 1이상)
        if(df_d.iloc[i].iloc[0]> 0 and df_d.iloc[i+1].iloc[0] <= 0):
            list.append(i+1)
    for i in range(len(df_open)):
        get_data = []
        if(i in list):
            get_data.append(df_open.iloc[i])
            get_data.append(df_high.iloc[i])
            list_data.append(get_data)
            continue
        get_data.append(np.NaN)
        get_data.append(np.NaN)  
        list_data.append(get_data)
    data = DataFrame(list_data, columns=['open','high'])
    return data

def get_line_df(start,end,df_d, Trend):
    if Trend:
        line_Df = data_uptrend(start,end,df_d,df['open'],df['low'])
        if len(line_Df.dropna().index) <= 1:
            return print("***데이터가 적어 분석할 수 없습니다.***") 
    else:
        line_Df = data_downtrend(start,end, df_d,df['open'], df['high'])
        if len(line_Df.dropna().index) <= 1:
            return print("***데이터가 적어 분석할 수 없습니다.***")
    return line_Df

line_Df = get_line_df(1, end_index, Df_d, Trend)

#DataFrame인 line_Df를 List로 변경
def get_line_list(Trend, line_Df):
    x = []
    y = []
    line_Df = line_Df.dropna()
    if Trend:
        for i in range(len(line_Df.index)):
            x.append(line_Df.index[i]) 
            y.append(line_Df['open'].iloc[i])
        for i in range(len(line_Df.index)):
            x.append(line_Df.index[i]) 
            y.append(line_Df['low'].iloc[i])  
        #print(x,y)
    else:
        for i in range(len(line_Df.index)):
            x.append(line_Df.index[i]) 
            y.append(line_Df['open'].iloc[i])
        for i in range(len(line_Df.index)):
            x.append(line_Df.index[i]) 
            y.append(line_Df['high'].iloc[i])  
        #print(x,y)
    #x = list(map(float,x))
    #y = list(map(float,y))
    return x , y

X, Y = get_line_list(Trend, line_Df)
print(X,Y)

#tensorflow 선형회귀를 통하여 지지선/저항선 도출
def Linear_Regression_1(X,Y):
    import tensorflow.compat.v1 as tf  
    tf.disable_v2_behavior()

    W = tf.Variable(tf.random.normal([1]), name = 'weight')
    b = tf.Variable(tf.random.normal([1]), name = 'bias')

    hypothesis = X * W + b

    cost = tf.reduce_mean(tf.square(hypothesis - Y))
    optimizer = tf.train.GradientDescentOptimizer(0.0001) #학습률: 0.0001
    train = optimizer.minimize(cost)

    sess = tf.Session()
    sess.run(tf.global_variables_initializer())
    
    for step in range(1000000): #학습 횟수: 1000000
        sess.run(train)
        if(step % 100000 == 0):
            print(step, sess.run(cost), sess.run(W), sess.run(b))
    
    return sess.run(W),sess.run(b)


#Tensorflow 2 사용 버전
def Linear_Regression_2(X,Y):
    import tensorflow as tf
    
    W = tf.Variable(tf.random.normal([1]), name = 'weight')
    b = tf.Variable(tf.random.normal([1]), name = 'bias')

    hypothesis = X * W + b

    cost = tf.reduce_mean(tf.square(hypothesis - Y))
    sgd = tf.keras.optimizers.SGD(learning_rate = 0.01)  #확률적 경사하강법 사용
    
    model = tf.keras.models.Sequential()
    model.add(tf.keras.layers.Dense(1,input_dim = 1)) #입력 1차원, 출력 1차원
    
    model.compile(loss='mean_squared_error',optimizer=sgd)
    
    model.fit(X,Y,epochs = 2000)
    
    weights = model.layers[0].get_weights()[0]
    bias = model.layers[0].get_weights()[1]
    return model.predict(X)


W,b = Linear_Regression_1(X,Y)
print(type(W[0]))
#지지선, 저항선 그래프
def f(w,b):
    x = np.arange(0,101,1)
    y = x*w + b
    return y

#그래프 그리기
fig = plt.figure(figsize=(12, 7))
ax2 = fig.add_subplot(1,1,1)
index = df.date.astype('str')
ax2.xaxis.set_major_locator(ticker.MaxNLocator(12))

ax2.plot(index, f(W,b),label='line')
#ax2.plot(index, Linear_Regression_2(X,Y), label = 'line')
mpl_finance.candlestick2_ohlc(ax2, df['open'], df['high'], df['low'], df['close'], width=0.5, colorup='r', colordown='b')
#ax2.plot(index, df['close'].rolling(window=1).mean(), label='MA1')
ax2.plot(index, line_Df ,'ro' ,label='plot')
ax2.legend(loc=1) 
min = df['low'].min()-1 
max = df['high'].max()+1
plt.ylim([min, max]) 
ax2.set_xlabel("Date")
ax2.set_ylabel("Price") 
plt.xticks(rotation = 45)
plt.grid()
plt.show()
```

4. 결론: 
 
   계획했던 것과 다르게 분석 도구를 만들며 많은 차질이 발생했다. 초기 추세선을 파악하여 지지선 또는 저항선을 그린 다음 돌파하는 시점으로 변화한 추세에 맞게 차트를 분석하게 하여 자동으로 불러온 차트에 빠진 구간 없이 지지선 또는 저항선을 그리려 했다. 하지만, 101개의 캔들만으로는 초기의 추세를 알기 어려웠다. 이전의 추세를 알기 위해서는 또 그 이전의 차트를 불러와야 했기 때문이다. 초기의 추세가 정해지면 그 다음 돌파 지점을 알 수 있으므로 초기의 추세를 강제적으로라도 정해주려하였다. 그러나 예상했던 것과 다르게 다른 변수도 생겼다. 초기의 차트가 상승하는 차트의 아주 끝 부분 이거나 하락하는 차트의 아주 끝 부분이면 상승세인지 하락세인지 알기 어려웠다. 코인 시장에는 피보나치 되돌림이라하여 올라갈 때 또는 내려갈 때 약간의 반등이 있다. 이 약간의 반등 부분을 보고 상승세인지 하락세인지 판단하는 것은 거의 불가능에 가까웠다. 이에 계획했던 것과는 다르지만 사용자가 직접 구간과 추세를 지정하여 차트를 분석할 수 있게 구현하였다. 
   
   프로젝트를 진행해 나감에 있어 걸림돌이 된 부분이 더 있었다. Tensorflow를 이용한 선형회귀에서 초기에 가중치(W)값과 편향(b)값을 정규분포 랜덤변수로 지정했다. 이 값들은 학습을 진행해 나감으로써 비용(cost)가 최소가 되는 값으로 수렴해간다. 학습률이 높을수록 W와 b 값이 큰 범위로 변화하는데 이는 주어진 데이터에 따라 수렴하지 않고 지나치게 높은 값을 가지게 되어 발산할 수도 있다. 반대로 학습률이 낮을수록 학습횟수에 따라 수렴하는 값에 천천히 다가간다. 프로젝트를 진행하며 117개의 거래소의 차트를 모두 불러올 수 있게 하였다. 모든 차트에서 원하는 만큼 데이터를 뽑아낼 수 없었고 각각의 차트마다 성향이 다 달랐다. 이에 따라 어느 차트는 선형회귀 학습이 잘 이루어져 수렴하는 값을 반환하는 반면 어느 차트는 cost가 nan이나 inf 값을 도출하며 가중치와 편향값을 구할 수 없게 되었다. 각 코인 차트마다 학습률과 학습횟수를 매번 바꿀 수 없기 때문에 낮은 학습률과 높은 학습횟수를 통하여 오류가 발생하지 않도록 하였다. 
   
   결론적으로 분석이 잘 되는 차트에는 저항선과 지지선이 예상한대로 잘 그어졌다. 돌파 지점에서 추세가 변화하는 현상도 확인할 수 있었다. 하지만, 언제나 저항선 지지선이 들어맞지는 않았다. 저항선 지지선은 말 그대로 분석 도구 중 가장 기본이였고 더 많은 분석 도구 또는 기술을 필요로 한다. 또한, 차트에서 분석에 필요한 극솟값이나 극댓값 등 데이터의 전처리 과정이 더 정교하게 이루어 졌다면 차트 분석이 더 정확했을 것이라고 생각한다.
